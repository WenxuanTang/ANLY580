```python
import json
import pandas as pd
import translators as ts
import re
```

### Step 1: Use google's API to translate the Chinese text.

```python
# read the data
with open("/content/clothing_1w_1.json") as file:
  df = json.load(file)

# print the length
print('Length of', input, 'is', len(df))

# print an example
print('Print an example:')
print(df[list(df.keys())[0]])

# store all the IDs
IDs = list(df.keys())

# create an empty dict to store results
res = {}
count = 0
for id in IDs: # decompose version: IDs[2500:5000]
  res[id] = {}
  res[id]['tgt'] = [ts.google(''.join(df[id]['tgt']))] # list to str, then str to list
  res[id]['cate'] = ts.google(df[id]['cate'])
  res[id]['src'] = ts.google(df[id]['src'])

  # update the count
  count += 1
  print(count)

# write into json file
with open("/content/ENG_clothing_1w_1.json", 'w') as json_file:
  json.dump(res, json_file)
```

### Step 2: For those texts that are not entirely translated into English, the following cells translate them again.

```python
with open('/content/ENG_clothing_1w_1.json') as f:
  data = json.load(f)
```

```python
IDs = list(data.keys())
```

#### Update for 'src'

```python
zhPattern = re.compile(u'[\u4e00-\u9fa5]+')

for id in IDs: # decompose version: IDs[5000:7500]
  text = data[id]['src'] # only process src

  # if Chinese still exists
  if zhPattern.search(text):
    print('Wrong Version:')
    print(data[id])

    # find index
    idx_list = [(m.start(0), m.end(0)) for m in zhPattern.finditer(text)]

    # translate
    start = 0
    res = ''
    for i in idx_list:
      res += text[start:i[0]]
      res += ts.google(text[i[0]:i[1]])
      start = i[1]

    last = idx_list[-1][1]
    res += text[last:]

    # update the dict
    data[id]['src'] = res

    print('Correct Version:')
    print(data[id]['src'])
```

#### Update for 'tgt'

```python
zhPattern = re.compile(u'[\u4e00-\u9fa5]+')

# test the first 100 IDs
for id in IDs[5000:7500]:
  text = data[id]['tgt'][0] # only process src
  
  # if Chinese still exists
  if zhPattern.search(text):
    print('Wrong Version:')
    print(data[id]['tgt'])

    # find index
    idx_list = [(m.start(0), m.end(0)) for m in zhPattern.finditer(text)]

    # translate
    start = 0
    res = ''
    for i in idx_list:
      res += text[start:i[0]]
      res += ts.google(text[i[0]:i[1]])
      start = i[1]

    last = idx_list[-1][1]
    res += text[last:]

    # update the dict
    data[id]['tgt'] = [res]

    print('Correct Version:')
    print(data[id]['tgt'])
```
