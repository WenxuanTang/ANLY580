```python
import os
os.chdir('/content/drive/MyDrive/GU/580/Data/photo')
import torchvision
import torch
import numpy as np
import collections
import matplotlib.pyplot as plt
from tqdm import tqdm
```

### Resize and uniform format

```python
resizer = torchvision.transforms.Resize((600,600))
for folder in tqdm(os.listdir('img')[2:]):
  folder = os.path.join('img',folder)
  resized_path = os.path.join(folder, 'resized') 
  os.mkdir(resized_path)
  for file in tqdm(os.listdir(folder)):
      filename=file
      file=os.path.join(folder,filename)
      
      img=torchvision.io.read_image(file)
      img=resizer(img)/255
      
      torchvision.utils.save_image(tensor=img,fp=os.path.join(resized_path,filename),format='png')
```
